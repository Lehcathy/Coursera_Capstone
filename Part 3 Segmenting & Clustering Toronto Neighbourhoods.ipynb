**Segmenting and Clustering Neighbourhoods in Toronto: Wikipedia Scraping**

Import libraries

```python
import numpy as np
import pandas as pd
import requests
import json

import bs4
from bs4 import BeautifulSoup

from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html

from geopy.geocoders import Nominatim 
from pandas.io.json import json_normalize
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
```

Scrape table from Wikipedia page

```python
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = requests.get(link).text
soup = BeautifulSoup (page, "lxml")

print(soup.title)
table = str(soup.table)
display_html(table, raw=True)
```

Convert html table to pandas dataframe

```python
df0=pd.read_html(table)
df=df0[0]
df.head(10)
```

Preprocess and clean data
1. Drop not assigned borough
2. Combine same postalcode neighbourhoods 
3. Replace not assigned neighbourhoods with borough names

```python
df1 = df[df.Borough!='Not assigned'] 

df2= df1.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

df2['Neighbourhood']=np.where(df2['Neighbourhood']=='Not assigned',df2['Borough'], df2['Neighbourhood'])

df2.head()
```

Get shape of dataframe

```python
df2.shape
```

**Segmenting and Clustering Neighbourhoods in Toronto: Neighbourhood Coordinates**

Import csv file

```python
coordinates = pd.read_csv('http://cocl.us/Geospatial_data')
coordinates
```

Merge tables based on postalcode/postcode

```python
coordinates.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(df2,coordinates,on='Postcode')
df3.head()
```

**Segmenting and Clustering Neighbourhoods in Toronto: Explore the Cluster**

Import libraries for mapping

```python
import numpy as np 
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json 
!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim 

import requests
from pandas.io.json import json_normalize

import matplotlib.cm as cm
import matplotlib.colors as colors

from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 
import folium 
```

Get rows from dataframe which include Toronto

```python
df4=df3[df3['Borough'].str.contains('Toronto', regex=False)]
df4.head()
```

Create map of Toronto using coordinates

```python
toronto_map = folium.Map([43.653963, -79.387207],zoom_start=10)
```

Add markers

```python
for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(toronto_map)
    
toronto_map
```

Cluster neighbourhoods with kmeans

```python
k=5
toronto_cluster = df4.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_cluster)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df4
```

Create map of Toronto with clusters

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
